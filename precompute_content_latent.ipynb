### Precomputing before training

It is to reduces GPU/CPU loads, since I have plenty of storage to spare.

This will avoid LazyImport problem..

First, it will load ContentVec, extract content_latent, save

Second, reset and do the same for speaker_latent



```python
import torch
from contentvec import *
from vctk_dataloader import VCTKDataset

device = get_default_device()
print("Torch CUDA available:", torch.cuda.is_available())
print("Using device:", device)

content_encoder = load_content_encoder(device=device)
print("Content Encoder Model:", type(content_encoder).__name__)

dataset = VCTKDataset(root="datasets/vctk/wav48_silence_trimmed", mic="mic1")
sample = dataset[0]
print("Dataset size:", len(dataset))
print("Speaker ID:", sample["speaker_id"])
print("Utterance ID:", sample["utterance_id"])
print("Mic:", sample["mic"])
print("Waveform shape:", sample["waveform"].shape)
print("Sample rate:", sample["sample_rate"])
print("Path:", sample["path"])
```

```python
content_latent = encode_content(content_encoder, sample["waveform"], device=device)
print("Content Latent Shape:", content_latent.shape)
print("Content Latent Device:", content_latent.device)
```
